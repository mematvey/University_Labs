```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Выберите не менее 5 значений радиуса круга $r∈(0,a].$

```python
a = 5
r_list = [i for i in range(1, 6)]
```

2(a) Рассчитайте истинную геометрическую вероятность p как отношение соответствующих площадей.

```python
# Функция для вычисления истинной геометрической вероятности
def true_probability(r, a):
    return (np.pi * r**2) / (4 * a**2)

p_true_list = list(map(
    true_probability,
    r_list,
    [a]*len(r_list)
))

print(p_true_list)
```

2(b) При помощи генератора случайных чисел разместите точки в Ω и определите их принадлежность A(r).

```python
n_points = 10000

def generate_points(n, a):
    return np.random.uniform(-a, a, (n, 2))

points = generate_points(n_points, a)

# Функция для определения кол-ва точек, принадлежащих кругу
def points_in_circle(points, r):
    return np.sum(points[:, 0]**2 + points[:, 1]**2 <= r**2)

n_in_circle_list = list(map(
    points_in_circle,
    [points]*len(r_list),
    r_list
))

# Определяем долю точек, попавших в круг
p_est_list = list(map(
    lambda x, y: x / y,
    n_in_circle_list,
    [n_points]*len(r_list)
))

print(p_est_list)
```

2(c) Постройте график $\hat{p}(n)$, где $\hat{p}$ — доля точек, которые попали в круг, а n — количество точек

```python
fig, axs = plt.subplots(5, 1, figsize=(15, 10))

for i in range(5):
    n_values = np.arange(1, n_points + 1)
    # Доля точек, попавших в круг
    p_hat_values = np.cumsum(points[:, 0]**2 + points[:, 1]**2 <= r_list[i]**2) / n_values 

    axs[i].plot(n_values, p_hat_values, label='$\hat{p}(n)$')
    axs[i].axhline(y=p_true_list[i], color='r', linestyle='--', label='True $p$')
    axs[i].set_title(f'Radius r = {r_list[i]}')

plt.legend()
plt.xlabel('Number of points $n$')
axs[2].set_ylabel('$\hat{p}(n)$')
plt.show()
```

2(d) Постройте график $ε(n)=|\hat{p}(n)−p|$ для анализа изменения ошибки оценки геометрической вероятности.

```python
fig, axs = plt.subplots(5, 1, figsize=(15, 10))

for i in range(5):
    n_values = np.arange(1, n_points + 1)
    # Доля точек, попавших в круг
    p_hat_values = np.cumsum(points[:, 0]**2 + points[:, 1]**2 <= r_list[i]**2) / n_values 
    # Модуль разницы ожидаемой и реальной доли точек, попавших в круг
    epsilon_values = np.abs(p_hat_values - p_true_list[i]) 

    axs[i].plot(n_values, epsilon_values, label='$\epsilon(n)$')
    axs[i].axhline(y=0, color='r', linestyle='--', label='Target $p$')
    axs[i].set_title(f'Radius r = {r_list[i]}')

plt.legend()
plt.xlabel('Number of points $n$')
axs[2].set_ylabel('$\epsilon(n)$')
plt.show()
```

3. Зафиксируйте последовательность {$ε_i$}. Для каждого значения r вычислите необходимое количество случайных точек N, необходимых для достижения данной точности $ε_i$. Постройте графики $N(ε)$.

```python
# Значения точности
epsilon_values = [10**-i for i in range(1, 6)]

fig, axs = plt.subplots(5, 1, figsize=(15, 10))

for i in range(5):
    n_values = np.arange(1, n_points + 1)
    # Доля точек, попавших в круг
    p_hat_values = np.cumsum(points[:, 0]**2 + points[:, 1]**2 <= r_list[i]**2) / n_values
    # Модуль разницы ожидаемой и реальной доли точек, попавших в круг
    epsilon_values_plot = np.abs(p_hat_values - p_true_list[i]) 

    N_values = []
    for epsilon in epsilon_values:
        # Индекс первого элемента, удовлетворяющего условию
        N = np.argmax(epsilon_values_plot <= epsilon) + 1
        N_values.append(N)

    axs[i].plot(epsilon_values, N_values, 'o-', label='$N(\epsilon)$')
    axs[i].axhline(y=0, color='r', linestyle='--', label='Target $p$')
    axs[i].set_title(f'Radius r = {r_list[i]}')

plt.legend()
plt.xlabel('$\epsilon$')
axs[2].set_ylabel('$N(\epsilon)$')
plt.show()
```
